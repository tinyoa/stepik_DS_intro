Профессия дата сайентиста пользуется успехом, и вместе с популярностью растет количество различных специализаций. Очень заметным стало отделение технологических профессий от бизнес-направления:

- Data Engineer(технологическая сторона): работа с данными(хранение, очистка, подготовка к обработке, разарботка функций и алгоритмов, совместно с аналитиками,... - 75% технологий и 25% бизнеса

- Analytics Translator(бизнес сторона): выбор источников и сбор данных, разработка функций совместно с инженерами, поиск новых бизнес-идей на основе данных и их дальнейшая реализация - 75% бизнеса и 25% технологий

С Data Engineer все более или менее понятно, а специализация Analytics Translator относительно новая. Чем именно они занимаются?

1. Переводят математические выводы на человеческий язык, создавая более доверительное отношение к технологии в целом. Для этого крайне полезно уметь визуализировать данные и очень поможет опыт работы с когнитивными искажениями.

2. Дружат бизнес с числовыми данными:
- расставляют приоритеты(нужно знать бизнес процессы и аналитические методы, понимать что можно реализовать, что нет)
- определяют какие именно данные нужны для обработки(потребуется и понимание динамики бизнес-процессов, и знание инструментов обработки данных)
- ищут эффективные решения с результатами в наиболее понятной форме(для работы с сообществом бизнес-пользователей важно умение объяснять аналитику простыми словами)
- создают простые практичные рекомендации для бизнеса и следят, чтобы идеи, возникшие в процессе обработки данных, были реализованы до конца.

3. Выстраивают организацию работы с данными на всех уровнях, чтобы каждый имел доступ к нужной информации, и решения всегда принимались с учетом аналитики.
